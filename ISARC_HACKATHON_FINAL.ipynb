## **COMPARING THE COMPRESSIVE STRENGTH OF CONCRETE MIXTURES USING DIFFERENT KINDS MACHINE LEARNING ALGORITHMS AND EVALUATION METRICS.**

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, median_absolute_error
from sklearn.impute import SimpleImputer
import lightgbm as lgb
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from keras.wrappers.scikit_learn import KerasRegressor


# Load the data here
data = pd.read_csv('/content/concrete_data.csv', encoding='cp1252')

# Feature Engineering - Add relevant features

# Deleting the rows with NULL values for target variables
data.dropna(subset=['Compressive\nstrength (f’c) (MPa)'], inplace=True)

# Splitting the data into input features (X) and target variables (y)
X = data.drop(['Compressive\nstrength (f’c) (MPa)','Compressive\nstrength tests','Elastic\nmodulus tests','Flexural\nstrength tests','Splitting\ntensile strength tests'], axis=1)
y = data['Compressive\nstrength (f’c) (MPa)']

# Using mean strategy for missing values in the input features
imputer = SimpleImputer(strategy='mean')
X = imputer.fit_transform(X)

# Feature selection using SelectKBest and f_regression
selector = SelectKBest(f_regression, k=5)
X_selected = selector.fit_transform(X, y)

# Splitting the data into training and testing sets, 80-20 Split
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, train_size=0.8, random_state=43)

# Define a function to create the neural network model
def create_neural_network():
    model = Sequential()
    model.add(Dense(64, activation='relu', input_shape=(X_train.shape[1],)))
    model.add(Dropout(0.2))
    model.add(Dense(64, activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(1))
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model

# Creating a KerasRegressor wrapper for the neural network model
neural_network = KerasRegressor(build_fn=create_neural_network, epochs=100, batch_size=32, verbose=0)

# Creating a list of algorithms with their hyperparameter grids
algorithms = [
    {
        'name': 'Linear Regression',
        'model': LinearRegression(),
        'param_grid': {}
    },
    {
        'name': 'Lasso Regression',
        'model': Lasso(),
        'param_grid': {
            'alpha': [0.01, 0.1, 1.0]
        }
    },

    {
        'name': 'Neural Network',
        'model': neural_network,
        'param_grid': {}
    },
    {
        'name': 'Ridge Regression',
        'model': Ridge(),
        'param_grid': {
            'alpha': [0.01, 0.1, 1.0]
        }
    },
    {
        'name': 'Elastic Net',
        'model': ElasticNet(),
        'param_grid': {
            'alpha': [0.01, 0.1, 1.0],
            'l1_ratio': [0.2, 0.5, 0.8]
        }
    },
    {
        'name': 'Random Forest',
        'model': RandomForestRegressor(),
        'param_grid': {
            'n_estimators': [100, 500, 1000],
            'max_depth': [None, 5, 10]
        }
    },
    {
        'name': 'Support Vector Regression',
        'model': SVR(),
        'param_grid': {
            'C': [1, 10, 100],
            'kernel': ['linear', 'rbf']
        }
    },
    {
        'name': 'LightGBM',
        'model': lgb.LGBMRegressor(),
        'param_grid': {
            'n_estimators': [100, 500, 1000],
            'learning_rate': [0.01, 0.1, 0.5],
            'max_depth': [3, 5, 7]
        }
    },
    {
        'name': 'Gradient Boosting',
        'model': GradientBoostingRegressor(),
        'param_grid': {
            'n_estimators': [100, 500, 1000],
            'learning_rate': [0.01, 0.1, 0.5],
            'max_depth': [3, 5, 7]
        }
    },

]

# Creating a list to store the evaluation metrics and predicted values for each algorithm
results = []

# Training and testing each algorithm
for algorithm in algorithms:
    model = algorithm['model']
    name = algorithm['name']

    # Hyperparameter tuning using GridSearchCV
    if 'param_grid' in algorithm:
        param_grid = algorithm['param_grid']
        grid_search = GridSearchCV(estimator=model, param_grid=param_grid, scoring='neg_mean_squared_error', cv=5)
        grid_search.fit(X_train, y_train)
        best_model = grid_search.best_estimator_
        model = best_model

    # Training the model
    model.fit(X_train, y_train)

    # Making predictions on the test set
    y_pred = model.predict(X_test)

    # Calculating evaluation metrics
    mse = mean_squared_error(y_test, y_pred)
    rmse = mean_squared_error(y_test, y_pred, squared=False)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    medae = median_absolute_error(y_test, y_pred)
    mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100

    # Storing the results
    results.append({'Algorithm': name, 'MSE': mse, 'RMSE': rmse, 'MAE': mae, 'R2 Score': r2, 'Median Absolute Error': medae, 'MAPE': mape, 'Predicted': y_pred, 'Actual': y_test})

# Creating a DataFrame to store the results
results_df = pd.DataFrame(results)

# Printing the results as evaluation metrics and predicted values --- line graph
for index, row in results_df.iterrows():
    print(f"Algorithm: {row['Algorithm']}")
    print(f"Mean Squared Error: {row['MSE']}")
    print(f"Median Absolute Error: {row['Median Absolute Error']}")
    print(f"Mean Absolute Percentage Error: {row['MAPE']}")
    print("Predicted Values:")
    for i in range(len(row['Predicted'])):
        print(f"Actual: {y_test.iloc[i]}, Predicted: {row['Predicted'][i]}")
    print()

    # Plotting actual vs predicted values as line graph
    plt.figure(figsize=(10, 6))
    plt.plot(y_test.index, y_test.values, label='Actual', marker='o')
    plt.plot(y_test.index, row['Predicted'], label='Predicted', marker='o')
    plt.xlabel('Index')
    plt.ylabel('Compressive Strength')
    plt.title(f"Actual vs Predicted Values for {row['Algorithm']}")
    plt.legend()
    plt.show()
    print()

    # Plotting feature importance for Random Forest(as this algorithm gives better accuaracy we focus more on this)
    if row['Algorithm'] == 'Random Forest':
        feature_importances = best_model.feature_importances_
        feature_names = selector.get_support(indices=True)
        feature_names = [data.columns[idx] for idx in feature_names]

        sorted_indices = feature_importances.argsort()
        sorted_names = [feature_names[idx] for idx in sorted_indices]

        plt.figure(figsize=(10, 6))
        plt.barh(range(len(feature_importances)), feature_importances[sorted_indices], tick_label=sorted_names)
        plt.xlabel('Feature Importance')
        plt.ylabel('Feature')
        plt.title('Feature Importance for Random Forest')
        plt.show()
        print()



        # Creating a correlation matrix for the selected features
        X_selected_df = pd.DataFrame(X_selected, columns=feature_names)
        correlation_matrix = X_selected_df.corr()

        plt.figure(figsize=(10, 6))
        sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
        plt.title('Correlation Matrix for Selected Features')
        plt.show()
        print()


# Plotting the bar graph to compare different algorithms based on MSE
plt.figure(figsize=(10, 6))
sns.barplot(x='Algorithm', y='MSE', data=results_df)
plt.xlabel('Algorithm')
plt.ylabel('Mean Squared Error')
plt.title('Comparison of Algorithms based on Mean Squared Error')
plt.xticks(rotation=45)
plt.show()
```

## **VISUAL REPRESENTATION OF COMPARISION OF ALGORITHMS USED WITH RESPECT TO DIFFERENT EVALUATION METRICS USED.**

```python
import matplotlib.pyplot as plt
import numpy as np

# Plotting the bar graph to compare different algorithms based on evaluation metrics
evaluation_metrics = ['MSE', 'RMSE', 'MAE', 'Median Absolute Error', 'MAPE', 'R2 Score']
bar_colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple', 'tab:brown']
algorithm_names = results_df['Algorithm'].values

plt.figure(figsize=(12, 8))

# Looping over each evaluation metric and plotting the bar plot
for i, metric in enumerate(evaluation_metrics):
    bars = plt.bar(
        x=np.arange(len(algorithm_names)) + i / len(evaluation_metrics),
        height=results_df[metric],
        color=bar_colors[i],
        width=1 / len(evaluation_metrics),
        label=metric
    )

    # Adding black borders around each column
    for bar in bars:
        bar.set_edgecolor('black')

# Adding accuracy percentage at the bottom of the bar plot
for i, row in results_df.iterrows():
    total_accuracy = np.mean([row[metric] for metric in evaluation_metrics if metric != 'R2 Score']) * 100
    plt.text(i, -0.02, f"{total_accuracy:.2f}%", ha='center', va='top', fontsize=10, color='black')

plt.xlabel('Algorithm', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.title('Evaluation Metrics for Different Algorithms', fontsize=14)
plt.xticks(ticks=np.arange(len(algorithm_names)), labels=algorithm_names, rotation=45, fontsize=10)
plt.legend(evaluation_metrics, fontsize=10)
plt.tight_layout()

# Adding grid lines
plt.grid(axis='y', linestyle='--')

# Adding a background color to the plot
plt.gca().set_facecolor('#f0f0f0')

# Adding a border around the plot
for spine in plt.gca().spines.values():
    spine.set_visible(True)
    spine.set_color('#999999')
    spine.set_linewidth(0.5)

# Changing the color and width of the ticks
plt.tick_params(axis='both', colors='#999999', width=0.5)

# Changing the color of the legend
plt.gca().get_legend().get_frame().set_edgecolor('#999999')

# Displaying the bar plot
plt.show()
```
